# Caso 1: Control de ventas diarias de una soda

## Problema
Una soda ubicada en San Pedro registra sus ventas de lunes a viernes. El programa debe calcular el total semanal, el promedio diario y clasificar el resultado según una meta.

## Entrada / datos
Use esta lista base:

```python
ventas_soda = [85000, 92000, 78000, 110000, 97000]
meta_semanal = 450000
```

## Proceso esperado

- Calcular total semanal.
- Calcular promedio diario.
- Comparar el total con la meta.
- Clasificar:
  - `Meta alcanzada` si el total es mayor o igual a la meta.
  - `Cerca de la meta` si el total alcanza al menos el 80% de la meta.
  - `Requiere atención` si no alcanza el 80%.

## Salida esperada
Debe mostrar total, promedio, porcentaje de cumplimiento y estado final.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|---|---|---|---|
| Caso normal | `[85000, 92000, 78000, 110000, 97000]`, meta `450000` | Meta alcanzada | |
| Caso límite | total exactamente igual a la meta | Meta alcanzada | |
| Caso especial | total menor al 80% de la meta | Requiere atención | |

````python
# Caso 1 - Escriba su solución aquí

ventas_soda = [85000, 92000, 78000, 110000, 97000]
meta_semanal = 450000

# Complete el análisis

venta_semanal = sum(ventas_soda)
promedio_diario = venta_semanal / len(ventas_soda)
porcentaje_cumplimiento = (venta_semanal / meta_semanal) * 100

if venta_semanal >= meta_semanal:
    estado = "¡Meta alcanzada!"
elif venta_semanal >= 0.8 * meta_semanal:
    estado = "Casi alcanzada"
else:
    estado = "Meta no alcanzada"

print (f"Venta semanal: ${venta_semanal}")
print (f"Promedio diario: ${promedio_diario:.2f}")
print (f"Porcentaje de cumplimiento: {porcentaje_cumplimiento:.2f}%")
print (f"Estado: {estado}")

````

# Caso 2: Asistencia a una actividad cultural

## Problema
Una actividad cultural en Costa Rica registra asistencia por bloques horarios. El programa debe calcular la asistencia promedio y clasificar el nivel de ocupación.

## Entrada / datos
Use esta lista base:

```python
asistencias = [120, 180, 210, 250, 230]
capacidad = 300
```

## Proceso esperado

- Recorrer la lista con un ciclo `for`.
- Calcular asistencia total y promedio.
- Contar cuántos bloques superaron el 70% de capacidad.
- Clasificar el promedio de ocupación:
  - `Alta ocupación` si el promedio es mayor o igual al 80% de la capacidad.
  - `Ocupación media` si el promedio es mayor o igual al 50%.
  - `Baja ocupación` si el promedio es menor al 50%.

## Salida esperada
Debe mostrar total de asistentes, promedio, bloques sobre 70% y estado de ocupación.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|---|---|---|---|
| Caso normal | `[120, 180, 210, 250, 230]`, capacidad `300` | Ocupación media o alta según cálculo | |
| Caso límite | promedio exactamente 80% de capacidad | Alta ocupación | |
| Caso especial | lista con asistencias muy bajas | Baja ocupación | |

````python
# Caso 2 - Escriba su solución aquí

asistencias = [120, 180, 210, 250, 230]
capacidad = 300

# Complete el análisis
total_asistentes = 0
bloques_sobre_70 = 0
umbral_70 = 0.7 * capacidad

for asistencia in asistencias:
    total_asistentes += asistencia
    if asistencia > umbral_70:
        bloques_sobre_70 += 1

promedio = total_asistentes / len(asistencias)
porcentaje_ocupacion = (promedio / capacidad) * 100

if promedio >= 0.8 * capacidad:
    estado = "Alta ocupación"
elif promedio >= 0.5 * capacidad:
    estado = "Ocupación media"
else:
    estado = "Baja ocupación"

# Salida
print(f"Total de asistentes: {total_asistentes}")
print(f"Promedio por bloque: {promedio:.2f}")
print(f"Porcentaje de ocupación promedio: {porcentaje_ocupacion:.2f}%")
print(f"Bloques sobre el 70% de capacidad: {bloques_sobre_70}")
print(f"Estado: {estado}")
````

# Caso 3: Precios de productos de feria

## Problema
Una persona compara precios de un mismo producto en diferentes ferias del agricultor. El programa debe analizar los precios y detectar si hay mucha variación.

## Entrada / datos
Use esta lista base:

```python
producto = "Tomate"
precios = [900, 850, 1000, 950, 1200]
```

## Proceso esperado

- Calcular precio promedio.
- Identificar precio mínimo y máximo.
- Calcular rango de precios: `máximo - mínimo`.
- Clasificar la variación:
  - `Variación alta` si el rango es mayor o igual a 300.
  - `Variación moderada` si el rango es mayor o igual a 150.
  - `Variación baja` si el rango es menor a 150.
- Usar un ciclo para contar cuántos precios están por encima del promedio.

## Salida esperada
Debe mostrar producto, promedio, mínimo, máximo, rango, precios sobre promedio y clasificación.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|---|---|---|---|
| Caso normal | `[900, 850, 1000, 950, 1200]` | Variación alta | |
| Caso límite | rango exactamente `300` | Variación alta | |
| Caso especial | todos los precios iguales | Variación baja | |

````python
# Caso 3 - Escriba su solución aquí

producto = "Tomate"
precios = [900, 850, 1000, 950, 1200]

# Complete el análisis
promedio = sum(precios) / len(precios)
precio_minimo = min(precios)
precio_maximo = max(precios)
rango = precio_maximo - precio_minimo

precios_sobre_promedio = 0
for precio in precios:
    if precio > promedio:
        precios_sobre_promedio += 1

if rango >= 300:
    clasificacion = "Variación alta"
elif rango >= 150:
    clasificacion = "Variación moderada"
else:
    clasificacion = "Variación baja"

# Salida
print(f"Producto: {producto}")
print(f"Precio promedio: {promedio:.2f}")
print(f"Precio mínimo: {precio_minimo}")
print(f"Precio máximo: {precio_maximo}")
print(f"Rango de precios: {rango}")
print(f"Precios sobre el promedio: {precios_sobre_promedio}")
print(f"Clasificación: {clasificacion}")
````

# Caso 4: Reto integrador - Red EmprendeCR

## Problema
Una red de emprendimientos costarricenses desea analizar el desempeño semanal de varias sedes. Cada sede tiene nombre, provincia, tipo de negocio, ventas diarias y meta semanal.

Este reto integra aprendizajes de las semanas 1 a 7:

- Semana 1: comprensión del problema y algoritmo.
- Semana 2: variables y entrada-proceso-salida.
- Semana 3: decisiones con condicionales.

- Semana 5: indicadores para análisis de datos.
- Semana 6: listas y ciclos.
- Semana 7: listas de diccionarios, funciones y elección de estructuras.

## Datos base

Use esta lista de diccionarios:

```python
sedes = [
    {
        "nombre": "Soda San Pedro",
        "provincia": "San José",
        "tipo": "Alimentación",
        "ventas": [85000, 92000, 78000, 110000, 97000],
        "meta": 450000,
    },
    {
        "nombre": "Panadería Cartago",
        "provincia": "Cartago",
        "tipo": "Panadería",
        "ventas": [65000, 70000, 72000, 69000, 75000],
        "meta": 380000,
    },
    {
        "nombre": "Cafetería Liberia",
        "provincia": "Guanacaste",
        "tipo": "Cafetería",
        "ventas": [98000, 105000, 99000, 120000, 115000],
        "meta": 500000,
    },
]
```

## Requisitos mínimos

1. Crear una función `calcular_total(ventas)`.
2. Crear una función `calcular_promedio(ventas)`.
3. Crear una función `clasificar_sede(total, meta)`.
4. Recorrer todas las sedes con un ciclo `for`.
5. Imprimir un reporte por sede con:
   - nombre;
   - provincia;
   - total semanal;
   - promedio diario;
   - porcentaje de cumplimiento;
   - estado.
6. Crear un `set` con las provincias presentes.
7. Crear una lista de tuplas `(nombre, total)` para guardar el ranking base.
8. Identificar la sede con mayor venta total.

## Clasificación requerida

- `Meta alcanzada`: total mayor o igual a la meta.
- `Cerca de la meta`: total mayor o igual al 80% de la meta.
- `Requiere atención`: total menor al 80% de la meta.

## Salida esperada
El reporte debe ser claro y comprensible para una persona que no está viendo el código.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|---|---|---|---|
| Caso normal | datos base | Reporte para las 3 sedes | |
| Caso límite | una sede con total igual a la meta | Meta alcanzada | |
| Caso especial | una sede con ventas muy bajas | Requiere atención | |

````python
# Caso 4 - Reto integrador

sedes = [
    {
        "nombre": "Soda San Pedro",
        "provincia": "San José",
        "tipo": "Alimentación",
        "ventas": [85000, 92000, 78000, 110000, 97000],
        "meta": 450000,
    },
    {
        "nombre": "Panadería Cartago",
        "provincia": "Cartago",
        "tipo": "Panadería",
        "ventas": [65000, 70000, 72000, 69000, 75000],
        "meta": 380000,
    },
    {
        "nombre": "Cafetería Liberia",
        "provincia": "Guanacaste",
        "tipo": "Cafetería",
        "ventas": [98000, 105000, 99000, 120000, 115000],
        "meta": 500000,
    },
]


# 1. Defina las funciones

def calcular_total(ventas):
    return sum(ventas)


def calcular_promedio(ventas):
    return calcular_total(ventas) / len(ventas)


def clasificar_sede(total, meta):
    if total >= meta:
        return "Meta alcanzada"
    elif total >= 0.8 * meta:
        return "Cerca de la meta"
    else:
        return "Requiere atención"


# 2. Recorra las sedes y genere el reporte

provincias = set()
ranking_base = []

print("=" * 50)
print("REPORTE SEMANAL - RED EMPRENDECR")
print("=" * 50)

for sede in sedes:
    total = calcular_total(sede["ventas"])
    promedio = calcular_promedio(sede["ventas"])
    porcentaje_cumplimiento = (total / sede["meta"]) * 100
    estado = clasificar_sede(total, sede["meta"])

    provincias.add(sede["provincia"])
    ranking_base.append((sede["nombre"], total))

    print(f"\nSede: {sede['nombre']}")
    print(f"Provincia: {sede['provincia']}")
    print(f"Total semanal: {total}")
    print(f"Promedio diario: {promedio:.2f}")
    print(f"Porcentaje de cumplimiento: {porcentaje_cumplimiento:.2f}%")
    print(f"Estado: {estado}")

# 3. Muestre provincias presentes, ranking base y mejor sede

print("\n" + "=" * 50)
print("RESUMEN GENERAL")
print("=" * 50)

print(f"\nProvincias con sedes activas: {provincias}")

print(f"\nRanking base (nombre, total):")
for nombre, total in ranking_base:
    print(f"  - {nombre}: {total}")

mejor_sede = max(ranking_base, key=lambda tupla: tupla[1])
print(f"\nSede con mayor venta total: {mejor_sede[0]} (₡{mejor_sede[1]})")
````
