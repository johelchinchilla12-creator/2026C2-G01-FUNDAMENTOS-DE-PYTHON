# Problema #1: Calculadora de propina inteligente

## Problema

El programa solicita al usuario el total de una cuenta de restaurante y el porcentaje de propina que desea dejar. A partir de esos datos, calcula el monto de la propina y el total final a pagar. El programa debe manejar entradas inválidas (texto o números negativos) y mostrar los resultados con formato claro.

## Entrada

- Total de la cuenta (número decimal, por ejemplo: `45.50`).
- Porcentaje de propina deseado (número decimal, por ejemplo: `15` para 15%).

## Proceso

- Convertir ambas entradas a `float` usando `try-except` para capturar errores de conversión.
- Validar que los valores sean mayores que cero.
- Calcular la propina: `propina = total * (porcentaje / 100)`
- Calcular el total final: `total_final = total + propina`

## Salida esperada

- Mensaje con el total de la cuenta, el monto de la propina y el total final, formateado con dos decimales usando f-strings.
- Ejemplo: `"Propina (15%): $6.83 | Total a pagar: $51.33"`

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | total=45.50, propina=15% | Propina: $6.83, Total: $52.33 | Propina: $6.82, Total: $52.32 |
| Caso límite | total=0.01, propina=0.1% | Propina: $0.00, Total: $0.01 | Propina: $0.00, Total: $0.01 |
| Caso inválido | total="abc", propina=10% | Error: ingrese un número válido | Error: ingrese un número válido |


```python
# Problema #1: Calculadora de propina inteligente

try:
    total_str = input("Ingrese el total de la cuenta ($): ").strip()
    total = float(total_str)

    porcentaje_str = input("Ingrese el porcentaje de propina (%): ").strip()
    porcentaje = float(porcentaje_str)

    if total <= 0 or porcentaje < 0:
        print("Error: los valores deben ser mayores que cero.")
    else:
        propina = total * (porcentaje / 100)
        total_final = total + propina

        print(f"\n--- Resumen de la cuenta ---")
        print(f"Total de la cuenta : ${total:.2f}")
        print(f"Propina ({porcentaje:.1f}%)     : ${propina:.2f}")
        print(f"Total a pagar      : ${total_final:.2f}")

except ValueError:
    print("Error: ingrese un número válido (use punto como separador decimal).")
```

# Problema #2: Detector de número par/impar

## Problema

El programa solicita al usuario un número entero y determina si dicho número es par o impar. Debe manejar entradas no numéricas con un mensaje de error claro.

## Entrada

- Un número entero ingresado por el usuario (positivo, negativo o cero).

## Proceso

- Limpiar la entrada con `.strip()` y convertir a `int` usando `try-except`.
- Aplicar el operador módulo: `numero % 2`.
  - Si el resultado es `0`, el número es **par**.
  - Si el resultado es `1` (o `-1` para negativos), el número es **impar**.

## Salida esperada

- Un mensaje indicando si el número ingresado es par o impar.
- Ejemplo: `"El número 7 es IMPAR."`

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 8 | El número 8 es PAR. | El número 8 es PAR. |
| Caso límite | 0 | El número 0 es PAR. | El número 0 es PAR. |
| Caso inválido | "doce" | Error: ingrese un número entero válido. | Error: ingrese un número entero válido. |


```python
# Problema #2: Detector de número par/impar

try:
    numero_str = input("Ingrese un número entero: ").strip()
    numero = int(numero_str)

    if numero % 2 == 0:
        resultado = "PAR"
    else:
        resultado = "IMPAR"

    print(f"El número {numero} es {resultado}.")

except ValueError:
    print("Error: ingrese un número entero válido (sin decimales ni letras).")
```

# Problema #3: Clasificador de edad

## Problema

El programa solicita la edad de una persona y la clasifica en una de tres categorías: menor de edad (0–17), adulto (18–64) o adulto mayor (65 en adelante). Debe validar que la edad sea un número entero positivo y manejar entradas inválidas.

## Entrada

- Edad en años (número entero, mayor o igual a cero).

## Proceso

- Convertir la entrada a `int` con `try-except`.
- Validar que la edad sea >= 0 y <= 120 (rango razonable).
- Aplicar lógica `if/elif/else`:
  - `edad < 18` → Menor de edad
  - `18 <= edad <= 64` → Adulto
  - `edad >= 65` → Adulto mayor

## Salida esperada

- Mensaje con la clasificación de la edad ingresada.
- Ejemplo: `"Con 17 años, la persona es: MENOR DE EDAD."`

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 30 | Con 30 años: ADULTO. | Con 30 años: ADULTO. |
| Caso límite | 18 | Con 18 años: ADULTO. | Con 18 años: ADULTO. |
| Caso inválido | -5 | Error: la edad no puede ser negativa. | Error: la edad no puede ser negativa. |


```python
# Problema #3: Clasificador de edad

try:
    edad_str = input("Ingrese su edad en años: ").strip()
    edad = int(edad_str)

    if edad < 0:
        print("Error: la edad no puede ser negativa.")
    elif edad > 120:
        print("Error: ingrese una edad válida (0–120 años).")
    elif edad < 18:
        categoria = "MENOR DE EDAD"
        print(f"Con {edad} años, la persona es: {categoria}.")
    elif edad <= 64:
        categoria = "ADULTO"
        print(f"Con {edad} años, la persona es: {categoria}.")
    else:
        categoria = "ADULTO MAYOR"
        print(f"Con {edad} años, la persona es: {categoria}.")

except ValueError:
    print("Error: ingrese un número entero válido para la edad.")
```

# Problema #4: Validador de compra con descuento

## Problema

El programa solicita el nombre del cliente, el monto de la compra y si es cliente frecuente. Aplica un descuento según las siguientes reglas:
- Cliente frecuente con compra mayor a $100 → 20% de descuento.
- Cliente frecuente con compra de $100 o menos → 10% de descuento.
- Cliente no frecuente con compra mayor a $200 → 5% de descuento.
- En cualquier otro caso → sin descuento.

## Entrada

- Nombre del cliente (texto).
- Monto total de la compra (número decimal).
- Si es cliente frecuente (respuesta: `"si"` o `"no"`).

## Proceso

- Limpiar el nombre con `.strip().title()`.
- Convertir el monto a `float` con `try-except`.
- Normalizar la respuesta de cliente frecuente con `.strip().lower()`.
- Evaluar las condiciones usando `and` / `or` para determinar el porcentaje de descuento.
- Calcular el descuento y el total final.

## Salida esperada

- Resumen con nombre, monto original, porcentaje y monto de descuento aplicado, y total final.
- Ejemplo: `"Cliente: Ana López | Descuento (20%): $30.00 | Total: $120.00"`

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | nombre=Ana, monto=150, frecuente=si | Descuento 20%: $30.00, Total: $120.00 | Descuento 20%: $30.00, Total: $120.00 |
| Caso límite | nombre=Luis, monto=100, frecuente=si | Descuento 10%: $10.00, Total: $90.00 | Descuento 10%: $10.00, Total: $90.00 |
| Caso inválido | nombre=Pedro, monto=abc, frecuente=no | Error: ingrese un monto numérico válido. | Error: ingrese un monto numérico válido. |


```python
# Problema #4: Validador de compra con descuento

try:
    nombre = input("Ingrese el nombre del cliente: ").strip().title()

    if not nombre:
        print("Error: el nombre no puede estar vacío.")
    else:
        monto_str = input("Ingrese el monto de la compra ($): ").strip()
        monto = float(monto_str)

        if monto <= 0:
            print("Error: el monto debe ser mayor que cero.")
        else:
            frecuente_str = input("¿Es cliente frecuente? (si/no): ").strip().lower()

            if frecuente_str not in ("si", "no"):
                print("Error: responda 'si' o 'no' para cliente frecuente.")
            else:
                es_frecuente = frecuente_str == "si"

                # Determinar porcentaje de descuento
                if es_frecuente and monto > 100:
                    porcentaje_descuento = 20
                elif es_frecuente and monto <= 100:
                    porcentaje_descuento = 10
                elif not es_frecuente and monto > 200:
                    porcentaje_descuento = 5
                else:
                    porcentaje_descuento = 0

                descuento = monto * (porcentaje_descuento / 100)
                total_final = monto - descuento

                print(f"\n--- Resumen de compra ---")
                print(f"Cliente          : {nombre}")
                print(f"Monto original   : ${monto:.2f}")
                if porcentaje_descuento > 0:
                    print(f"Descuento ({porcentaje_descuento}%)   : -${descuento:.2f}")
                else:
                    print(f"Descuento        : sin descuento")
                print(f"Total a pagar    : ${total_final:.2f}")

except ValueError:
    print("Error: ingrese un monto numérico válido (use punto como separador decimal).")
```

# Conclusión

**¿Qué validación resultó más importante?**

La validación más importante fue el uso de `try-except` para manejar errores de conversión de tipo (`ValueError`). Sin este bloque, cualquier entrada no numérica detiene el programa con un error sin información útil para el usuario. Al capturar la excepción, el programa responde con un mensaje claro y comprensible, lo que mejora significativamente la experiencia del usuario y la robustez del código.

**¿Qué error se encontró durante las pruebas?**

En el Problema #4 se detectó que la condición para el descuento del 10% no se evaluaba correctamente cuando el monto era exactamente $100. Inicialmente se usó `monto < 100` en lugar de `monto <= 100`, lo que dejaba sin descuento a un cliente frecuente con compra exactamente de $100. La corrección fue ajustar el operador de comparación a `<=`.

**¿Cómo mejoró la limpieza del código la calidad del programa?**

El uso de `.strip()` para eliminar espacios innecesarios en la entrada del usuario evitó falsos errores de validación (por ejemplo, ` si ` fallando la comparación con `"si"`). El uso de `.lower()` permitió aceptar respuestas como `"SI"`, `"Si"` o `"sí"` de forma más tolerante. Por su parte, los f-strings con formato `:.2f` garantizan que todos los valores monetarios se muestren siempre con dos decimales, dando una presentación profesional y consistente. En conjunto, estas prácticas hacen el código más legible, predecible y fácil de mantener.
